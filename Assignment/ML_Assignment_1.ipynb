```python
import numpy as np
import matplotlib.pyplot as plt

# Feature (x):  Size of the House (in square feet)
# Label (y):    Price of the House (in thousands of dollars)

x = np.array([1500, 2000, 1200, 1800, 2200, 1600, 1300, 2100, 1400, 1900, 1700, 2500, 1100, 2300, 2000, 2400, 1700, 1600, 1900, 2200])
y = np.array([300, 450, 250, 350, 500, 320, 280, 430, 290, 380, 400, 550, 240, 520, 420, 530, 380, 340, 410, 490])

# Function to calculate coefficients
def calculate_coefficients(x, y):
    num = np.size(x)
    
    # mean of x and y
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    
    # calculating cross-deviation and deviation about x
    SS_xy = np.sum(y*x) - num * mean_y * mean_x
    SS_xx = np.sum(x*x) - num * mean_x * mean_x
    
    # calculating regression coefficients
    coefficient_1 = SS_xy / SS_xx
    coefficient_0 = mean_y - coefficient_1 * mean_x
    return (coefficient_0, coefficient_1)

# Function to plot regression line
def plot_regression(x, y, coefficients):
    plt.scatter(x, y, color="m", marker="o", s=30)
    y_pred = coefficients[0] + coefficients[1] * x
    plt.plot(x, y_pred, color="g")
    plt.xlabel('Size of the House (sq. ft)')
    plt.ylabel('Price of the House (in 1000$)')
    plt.show()

# Estimate coefficients
coefficients = calculate_coefficients(x, y)

# Print coefficients
print(f"Estimated coefficients:\ncoefficient_0 = {coefficients[0]}\ncoefficient_1 = {coefficients[1]}")

# Plot regression line
plot_regression(x, y, coefficients)
```
